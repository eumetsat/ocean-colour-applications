Welcome to the MSOC data preparation code module, **fetch_my_data**. If you have not used Jupyter Notebook before, we suggest that you review the <a href="../../working-with-python/Intro_to_Python_and_Jupyter.ipynb" target="_blank">Introduction to Python and Project Jupyter<a> notebook in the *working-with-python* submodule folder or the <a href="https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Running%20Code.html" taget="_blank">"Running code"</a> section of the Jupyter Notebook Read the Docs.

The purpose of this code is to retrieve the data you need to run the notebooks in the MSOC course directory from the WEkEO object storage. Below, you can select the components you would like to download. You have the following options:

* Preprepared: this is the preprepared data you will need to run all of the Jupyter Notebooks provided with the MSOC course
* ROI1: this is the example data used to support the MSOC mini-project regions of interest. There are six to choose from (ROI1, ROI2, ROI3 etc).

You can prepare you components list as follows, for example:

`components = ["Preprepared"]` <br>
or <br>
`components = ["Preprepared", "ROI1", "ROI2"]`

The `Preprepared` data volume is 20 Gb. Each ROI is approximately 10 Gb. **The total volume of all options is 81Gb!**

By default, you should have activated the **cmts_ocean_colour_applications** environment.

```python
components = ["Preprepared"]
```

Once you have selected your components, you can run the cell below

```python
import s3fs
import os

fs = s3fs.S3FileSystem(
    anon=True,
    client_kwargs={"endpoint_url": "https://s3.waw3-2.cloudferro.com"}
)

for component in components:

    paths = fs.glob(f"wekeo/MSOC_data/{component}/**/*.*")
    total_size = 0
    sizes = {}
    
    for p in paths:
        info = fs.info(p)
        size = info["size"]
        sizes[p] = size
        total_size += size
    
    print(f"Total size for **{component}**: {total_size / 1e9:.2f} GB across {len(paths)} files")

    print(f"Downloading {component}")
    fs.get(
        f"s3://wekeo/MSOC_data/{component}/",   # remote prefix
        f"./Data/{component}",        # local destination
        recursive=True
    )
    print(f"{component} download complete")
```

Once you are happy with the data you have downloaded you can carry on with the rest of the notebooks in the MSOC directory.
