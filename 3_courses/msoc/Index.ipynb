# Index for the Multi-Sensor Ocean Colour Course module

### About this module

This software was developed by the MSOC training team to support the 2025 IOCCG/Copernicus funded Multi-sensor Ocean Colour course. The course, which is coordinated by EUMETSAT, in collaboration with NASA and ESA was held in Darmstadt, Germany between Dec 7th and Dec 11th, 2025, following the IOCS meeting. The tools and workflows included here reflect work performed by EUMETSAT under its Copernicus Marine Training Service contract, and by NASA in support of Pace Hack Week and other training activities.

For any questions about this repository, please contact <a href=mailto:training@eumetsat.int target="_blank">training@eumetsat.int</a>.

<hr>

### Module outline

The **msoc** module consists of a collection of Python-based Jupyter Notebooks design to demonstrate single- and multi-sensor ocean colour workflows from level-1 to level-4, with and focus on Copernicus Sentinel-3 OLCI, Copernicus Sentinel-2 MSI, NASA PACE OCI and the Copernicus Marine Service. The module will introduce you to:

* methods for accessing and downloading single- and multi-sensor ocean colour products
* approaches to atmospheric correction
* biooptical algorithm typology and selection
* multi-sensor synergy and merging
* product uncertainty and validation

<hr>

### How to use this material

This module is based around a series of <a href="https://jupyter.org/" target="_blank">Jupyter Notebooks</a>. These support high-level interactive learning by allowing us to combine code, text descriptions and data visualisations. If you have not worked with Jupyter Notebooks before, please look at the

<a href="../../working-with-python/Intro_to_Python_and_Jupyter.ipynb" target="_blank">Introduction to Python and Project Jupyter</a> module to get a short introduction to their usage and benefits. Refer to this module's **README** for further information on how to set up Jupyter.

There is a suggested order in which to access the notebooks (working through each day), and pre-requisites where applicable are listed at the top of each notebook. However, not all notebooks may be relevant for every user, so we have tried to make each one as self contained as possible. 

<div class="alert alert-success" role="alert">

#### Activating the correct environment

You must ensure that you have activated the correct environment (**cmts_ocean_colour_applications**) to use these notebooks. You can check that this has been loaded properly using the <a href="test_my_code.ipynb" target="_blank">"Test my code"</a> environment checker.

</div>

<div class="alert alert-success" role="alert">

#### Fetching supporting data:

While this repository contains all of the code used in the 2025 iteration of the MSOC course, it does not contain all of the supporting data, due its size. In total, there are 81 Gb of support products, however you do not require all of this to work with the examples. The supporting data is divided into two parts:

* **Preprepared data**: This is stored in the "Preprepared" folder and contains everything you need to run the scripts included in this notebook
* **ROI data**: This is stored in the "ROI*" folders and spectral_locations.md file and contains all of the example data used to support mini-projects focussed on a specific region of interest throughout the course. This is entirely optional if you are not using this code in the context of a specific course.

Both sets of data are stored in object storage on WEkEO and can be downloaded using the <a href="fetch_my_data.ipynb" target="_blank">"Fetch my data"</a> data preparation notebook.
</div>

### Learning outcomes

From this module, you can expect to learn:
* how to access, download and work with single-sensor and multi-sensor ocean colour data sets
* understand the advantages, consequences and limitations associated with differing approaches to atmospheric correction
* appreciate the complexity of selecting bio-optical algorithms for ocean colour retrievals
* be familiar with the key considerations needed for multi-sensor synergy and harmonisation
* know how to assess and apply product flags, determine product quality, and construct match-up validation workflows
* gain experience of operational satellite data production
