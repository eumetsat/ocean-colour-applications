## Exploring Sentinel-3 OLCI ocean colour products

### Learning outcomes¶
At the end of these notebooks you will know;

* What ocean colour variables are available in level-2 OLCI products from EUMETSAT
* what recent changes have been made to this product suite
* how to work with these products
  
### Outline

The Ocean and Land Colour Instrument (OLCI) aboard the Copernicus Sentinel-3 satellites provides frequent, high-quality observations of ocean colour, supporting the study and monitoring of marine ecosystems, water quality, and biogeochemical processes from regional to global scales. Its broad swath, high radiometric sensitivity, and 21 spectral bands—including bands designed specifically for aquatic applications—make OLCI particularly well suited to observing both open-ocean and optically complex coastal waters. 

The marine data stream from OLCI, produced by EUMETSAT, provides a range of ocean-colour products derived from these observations, including remote-sensing reflectance (as of baseline collection 004), chlorophyll-a concentration, total suspended matter, inherent optical properties (IOPs), and chlorophyll fluorescence. Together, these products allow users to move from measurements of the ocean’s spectral signal to estimates of key biological and optical properties of the marine environment.

In the notebooks below, we will explore some of these products.

For this pratical, we will use a series of notebooks that are included in the `learn-olci` repository. We have added this as a submodule in the root folder of this repository.

<div class="alert alert-block alert-danger">
In early 2026, EUMETSAT introduced a new operational processor for Sentinel-3 OLCI products; known as baseline collection BC004. This processor introduced a new atmospheric correction approach, a new chlorophyll algorithm, BRDF correction and an expanded product suite. You can find full information on the <b><a href="https://user.eumetsat.int/news-events/news/new-sentinel-3-olci-level-2-water-processing" target="_blank">EUMETSAT User Portal</a></b>. This processor introduces significant discontinuities with both the previous operational level 2 processor (BC003) and the current best available reprocessing (also BC003). In these modules, you can find notebooks suitable for working with both BC003 and BC004 level 2 products. Please take care to select the notebook that is appropriate for your baseline of interest. Level 1b products are not affected.
</div>
