# Index for the IOCCG Summer Lecture Series module

### About this module

These modules, which borrow heavily from other activities, were assembled to support the 2026 IOCCG Summer Lecure Series, coordinated by CNR, and hosted by CNR and the Venice International University (VIU). The tools and workflows included here reflect work performed by EUMETSAT under its Copernicus Marine Training Service contract, and by NASA in support of Pace Hack Week and other training activities.

For any questions about this repository, please contact <a href=mailto:training@eumetsat.int target="_blank">training@eumetsat.int</a>.

<hr>

### Module outline

The **ioccg-sls** module consists of a collection of Python-based Jupyter Notebooks design to demonstrate:

* methods for accessing and downloading single- and multi-sensor ocean colour products
* how to explore and use Copernicus Sentinel-3 Ocean Colour products
* how to validate OLCI and PACE products using in situ match-up analysis with the ThoMaS package

<hr>

### How to use this material

This module is based around a series of <a href="https://jupyter.org/" target="_blank">Jupyter Notebooks</a>. These support high-level interactive learning by allowing us to combine code, text descriptions and data visualisations. If you have not worked with Jupyter Notebooks before, please look at the

<a href="../../working-with-python/Intro_to_Python_and_Jupyter.ipynb" target="_blank">Introduction to Python and Project Jupyter</a> module to get a short introduction to their usage and benefits. Refer to this module's **<a href="https://gitlab.eumetsat.int/eumetlab/oceans/ocean-training/applications/ocean-colour-applications/-/blob/main/README.md?ref_type=heads" target="_blank">README</a>** for further information on how to set up Jupyter.

There is a suggested order in which to access the notebooks (working through each day), and pre-requisites where applicable are listed at the top of each notebook. However, not all notebooks may be relevant for every user, so we have tried to make each one as self contained as possible. 

<div class="alert alert-success" role="alert">

#### Activating the correct environment

You must ensure that you have activated the correct environment (**cmts_ocean_colour_applications**) to use these notebooks. You can check that this has been loaded properly using the <a href="test_my_code.ipynb" target="_blank">"Test my code"</a> environment checker.

### Learning outcomes

From this module, you can expect to learn:
* how to access, download and work with single-sensor and multi-sensor ocean colour data sets
* how to work with Copernicus Sentinel-3 OLCI ocean colour products
* how to assess and apply product flags, determine product quality, and construct match-up validation workflows
