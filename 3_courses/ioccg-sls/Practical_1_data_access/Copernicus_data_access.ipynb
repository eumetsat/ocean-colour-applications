### Outline

Ocean colour data from the Copernicus programme can be accessed through several complementary services. The EUMETSAT **<a href="https://data.eumetsat.int/" target="_blank">Data Store</a>** provides access to Sentinel-3 Ocean and Land Colour Instrument (OLCI) products, while the ESA **<a href="https://dataspace.copernicus.eu/" target="_blank">Copernicus Data Space Ecosystem (CDSE)</a>** is the primary source for Sentinel-2 Multi-Spectral Instrument (MSI) products. Higher-level ocean colour datasets, including Level-3 and Level-4 single-sensor and multi-sensor products, are available through the <a href="https://marine.copernicus.eu/" target="_blank">Copernicus Marine Service (CMEMS)</a>. For users seeking a single access points, the **<a href="https://wekeo.copernicus.eu/" target="_blank">WEkEO DIAS</a>** and **<a href="https://platform.destine.eu/" target="_blank">Destination Earth</a>** offer unified access to these Sentinel and CMEMS datasets through a common cloud-based platform.
